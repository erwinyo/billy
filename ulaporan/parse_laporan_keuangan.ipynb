```python
import pandas as pd
```

```python
SHEET_NAME = "JUNE"
df = pd.read_excel('Laporan Keuangan 2025.xlsx', sheet_name=SHEET_NAME)
df.head(10)
```

# Freedom Fund

```python
cells = []
for i in range(4, 66):
    cells.append([(i, 1), (i, 2), (i, 3)])  # (row, column) positions

values = []
for cell in cells:
    values.append([df.iat[r, c] for r, c in cell])

freedom_fund_rev_values = pd.DataFrame(values, columns=['DATE', 'DESCRIPTION', 'ISSUED'])
freedom_fund_rev_values = freedom_fund_rev_values.dropna(how='any')
freedom_fund_rev_values['FLOW'] = "IN"
freedom_fund_rev_values['WALLET'] = "freedom_fund"
freedom_fund_rev_values
```

```python
cells = []
for i in range(4, 66):
    cells.append([(i, 4), (i, 5), (i, 6)])  # (row, column) positions

values = []
for cell in cells:
    values.append([df.iat[r, c] for r, c in cell])

freedom_fund_exp_values = pd.DataFrame(values, columns=['DATE', 'DESCRIPTION', 'ISSUED'])
freedom_fund_exp_values = freedom_fund_exp_values.dropna(how='any')
freedom_fund_exp_values['FLOW'] = "OUT"
freedom_fund_exp_values['WALLET'] = "freedom_fund"
freedom_fund_exp_values
```

```python
freedom_wasting = pd.concat([freedom_fund_rev_values, freedom_fund_exp_values], ignore_index=True)
freedom_wasting
```
